```python
train = pd.read_csv('mnist_train.csv')
test = pd.read_csv('mnist_test.csv')
```

```python
x_train = train.iloc[0:5000, 1:]
x_test = train.iloc[5001:10000, 1:]
y_train = train.iloc[0:5000, 0]
y_test = train.iloc[5001:10000, 0]
x_train = x_train / 255.0
x_test  = x_test / 255.0
```

```python
mlp = MLPClassifier(hidden_layer_sizes=(100,), max_iter=20, alpha=1e-4,
                    solver='adam', verbose=1, random_state=1,
                    learning_rate_init=.1, activation = 'relu')
```

```python
mlp.fit(x_train, y_train)#adam, relu
```

```python
# --- Evaluate ---
print("Train accuracy:", mlp.score(x_train, y_train))
print("Test accuracy:", mlp.score(x_test, y_test))
```

```python
mlp = MLPClassifier(
    hidden_layer_sizes=(100,),
    activation='relu',
    solver='sgd',
    learning_rate_init=0.1,
    momentum=0.9,
    nesterovs_momentum=True,
    max_iter=20,
    verbose=1,
    random_state=1
)
mlp.fit(x_train, y_train)#sgd, relu|
```

```python
print("Train accuracy:", mlp.score(x_train, y_train))
print("Test accuracy:", mlp.score(x_test, y_test))
```

```python
# --- Adam + Identity Activation ---

# Normalize pixels (important)
x_train_norm = x_train
x_test_norm  = x_test



mlp_identity = MLPClassifier(
    hidden_layer_sizes=(100,),
    activation='identity',      # <--- Identity
    solver='adam',
    learning_rate_init=0.1,
    alpha=1e-4,
    max_iter=20,
    verbose=1,
    random_state=1
)

mlp_identity.fit(x_train_norm, y_train)

print("Identity Activation - Train Accuracy:", mlp_identity.score(x_train_norm, y_train))
print("Identity Activation - Test Accuracy:", mlp_identity.score(x_test_norm, y_test))
```

```python
#lbfgs, relu, max_iter 20 -> 50

mlp_lbfgs = MLPClassifier(hidden_layer_sizes=(100,), max_iter=50, alpha=1e-4,
                    solver='adam', verbose=1, random_state=1,
                    learning_rate_init=.1, activation = 'relu')


mlp_lbfgs.fit(x_train_norm, y_train)

print("LBFGS + ReLU - Train Accuracy:", mlp_lbfgs.score(x_train_norm, y_train))
print("LBFGS + ReLU - Test Accuracy:", mlp_lbfgs.score(x_test_norm, y_test))
```

```python
mlp = MLPClassifier(hidden_layer_sizes=(200,), max_iter=20, alpha=1e-4,
                    solver='adam', verbose=1, random_state=1,
                    learning_rate_init=.1, activation = 'relu')

mlp.fit(x_train, y_train)#adam, relu, hiddenlayersizes 100->200
```

```python
print("Train accuracy:", mlp.score(x_train, y_train))
print("Test accuracy:", mlp.score(x_test, y_test))
```
